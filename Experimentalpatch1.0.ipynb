```python
%matplotlib tk
from matplotlib import pyplot as plt 
import matplotlib.animation as animation
from matplotlib.animation import FuncAnimation
import numpy as np
# create empty lists for the x and y data
x = []
y = []
img = plt.imread("shrigma.jpg")

plt.style.use("dark_background")
fig, ax = plt.subplots()



def animate(i):
    if i <= 50:
        xylim = round(i)
    else:
        xylim = 50
        
    xlist = np.linspace(-xylim-i, xylim+i, 1000) # Create 1-D arrays for x,y dimensions
    ylist = np.linspace(-xylim-i, xylim+i, 1000) 
    X,Y = np.meshgrid(xlist, ylist) # Create 2-D grid xlist,ylist values
    F = np.sin(np.sin(X) + np.cos(Y)) - np.cos(np.sin((abs(X*Y))**(5/8)+np.cos(X)))
    ax.clear()
    ax.imshow(img,extent=[-xylim-i,xylim+i,-xylim,xylim])
    ax.contour(X, Y, F, [0], colors =["red"],linestyles = 'solid') #   x and y plotted using the contour plot
    
ani = FuncAnimation(fig, animate, interval=1,frames = 100)

```
